```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4],[10,15,34,23],label = 'BTC')
plt.xlabel('Days')
plt.ylabel('Price')
plt.title('Bitcoin Price')
plt.legend()
plt.show()
```

```python
market = [500,300,600,403]
price = [34,56,43,22]

plt.scatter(market,price,label = 'Stocks')
plt.xlabel('Market Cap')
plt.ylabel('Price')
plt.title('Stock Price')
plt.legend()
plt.show()
```

```python
days = [1,2,3,4]
sta_price = [20,40,56,43]
stb_price = [34,56,43,22]
plt.plot(days,sta_price,label = 'Stock A')
plt.plot(days,stb_price,label = 'Stock B')
plt.xlabel('Days')
plt.ylabel('Price')
plt.title('Stock Price')
plt.legend()
plt.show()
```
